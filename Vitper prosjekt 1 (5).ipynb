```python
from sklearn import preprocessing

def Sobel_filter(Alocal):
    n = len(Alocal)
    m = len(Alocal[0])
    
    
    
    g_x = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    g_y = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    
    X = np.zeros((n-2,m-2))
    Y = np.zeros((n-2,m-2))
    S = np.zeros((n-2,m-2))
    
    for i in range(2,n-1):
        for j in range (2,n-1):
            lokalePiksler = Alocal[i-1:i+2,j-1:j+2]
            X[i-1][j-1] = np.sum(g_x*lokalePiksler)
            Y[i-1][j-1] = np.sum(g_y*lokalePiksler)
            S[i-1][j-1] = np.sqrt((X[i-1][j-1])**2+(Y[i-1][j-1])**2)
            
    X = X/np.sum(X)
    Y = Y/np.sum(Y)
    S = S/np.sum((S))
            
    return X,Y,S

X,Y,S = Sobel_filter(I_matrise(pos_x,pos_y,M,N,0,100,n,0,100,n)) 

plt.pcolormesh(X)
plt.show()
plt.pcolormesh(Y)
plt.show()
plt.pcolormesh(S)
plt.show()

            
            
    
```
