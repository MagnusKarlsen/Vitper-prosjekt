```python
import numpy as np
import matplotlib.pyplot as plt
import timeit


dx = 1
dt = 1



def appendix1(M,pr):
    TilfeldigeTall = np.random.uniform(0,1,M)
    posisjon = np.zeros(M+1)
    tid = np.linspace(0, 1, M+1)
    
    
    
    for i in range (1,M):
        if TilfeldigeTall[i] < pr:
            posisjon[i] = posisjon[i-1] + dx
        else:
            posisjon[i] = posisjon[i-1] - dx
    return tid, posisjon




tid, posisjon = appendix1(120,0.5)

plt.plot(tid,posisjon)
plt.show()
tid, posisjon = appendix1(120,0.45)

plt.plot(tid,posisjon)
plt.show()
tid, posisjon = appendix1(120,0.55)


plt.plot(tid,posisjon)
plt.show()

timeit.timeit()

```
