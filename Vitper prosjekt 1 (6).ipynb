```python

def Sobel_filter(Alocal):
    
    
    #setter dimensjonene til sobe filteret lik dimensjonene til matrisen som tas inn
    n = len(Alocal)  
    m = len(Alocal[0])
    
    
    
    g_x = np.array([[1,0,-1],[2,0,-2],[1,0,-1]]) #definerer g_x
    g_y = np.array([[1,2,1],[0,0,0],[-1,-2,-1]]) #definerer g_y
    
    X = np.zeros((n-2,m-2)) #lager tom X matrise
    Y = np.zeros((n-2,m-2)) #lager tom Y matrise
    S = np.zeros((n-2,m-2)) #lager tom S matrise
    
    for i in range(2,n-1):
        for j in range (2,n-1):
            lokalePiksler = Alocal[i-1:i+2,j-1:j+2]   
            X[i-1][j-1] = np.sum(g_x*lokalePiksler)   #multipliserer et lokalt punkt/ element fra matrisen med g_x matrisen
            Y[i-1][j-1] = np.sum(g_y*lokalePiksler)   #multipliserer et lokalt punkt/ element fra matrisen med g_y matrisen
            S[i-1][j-1] = np.sqrt((X[i-1][j-1])**2+(Y[i-1][j-1])**2) #regner ut 
            
    X = X/np.sum(np.concatenate(X))  #normaliserer X matrisen
    Y = Y/np.sum(np.concatenate(Y))  #normaliserer Y matrisen
    S = S/np.sum(np.concatenate(S))  #normaliserer S matrisen
            
    return X,Y,S

X,Y,S = Sobel_filter(I_matrise(pos_x,pos_y,M,N,0,100,n,0,100,n))  #kaller på sobelfilter med I_matrisen for virrevandrerne som Alocal

plt.pcolormesh(X)  #plotter X matrisen
plt.show()
plt.pcolormesh(Y)  #plotter Y matrisen
plt.show()
plt.pcolormesh(S)  #plotter S matrisen
plt.show()

            
            
    
```
